```python
import pandas as pd

df = pd.read_csv("C:/Users/karth/OneDrive/Desktop/retail-forecast-segmentation-inventory/data/raw/Superstore.csv",
                 encoding="latin1")
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df[['Order Date', 'Ship Date']].head()
```

```python
df_model = df[['Order Date',
               'Customer ID',
               'Category',
               'Sub-Category',
               'Product ID',
               'Sales',
               'Quantity',
               'Profit']].copy()

df_model.head()
```

```python
df_model.isnull().sum()
```

```python
df_model['Year'] = df_model['Order Date'].dt.year
df_model['Month'] = df_model['Order Date'].dt.to_period('M').astype(str)

df_model[['Order Date', 'Year', 'Month']].head()
```

```python
output_path = "C:/Users/karth/OneDrive/Desktop/retail-forecast-segmentation-inventory/data/raw/Superstore.csv"
df_model.to_csv(output_path, index=False)
output_path
```
